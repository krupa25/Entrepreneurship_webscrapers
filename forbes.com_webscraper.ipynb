```python
from selenium import webdriver
import bs4
import pandas as pd
```

```python
driver = webdriver.Chrome('C:/Users/My PC/Downloads/chromedriver')
driver.get("https://www.forbes.com/entrepreneurs/#3bf66ce83035")
```

```python

uid = driver.find_elements_by_class_name("stream-item__text")

for a in uid:
    print(a.text)
    href = a.find_element_by_css_selector("a")
    print(href.get_attribute("href"))
    print("-"*70)
```

```python
import itertools
title = driver.find_elements_by_class_name("stream-item__title")
desc = driver.find_elements_by_class_name("stream-item__description")
date = driver.find_elements_by_class_name("stream-item__date")
author = driver.find_elements_by_class_name("byline__author-name")
link = driver.find_elements_by_class_name("stream-item__title")

for t,d,da,a,l in zip(title,desc,date,author,link):
    print(t.text)
    print(d.text)
    print(da.text)
    print("By "+a.text)
    print(l.get_attribute("href"))
    print("-"*70)
```

```python
title = driver.find_elements_by_class_name("stream-item__title")
desc = driver.find_elements_by_class_name("stream-item__description")
date = driver.find_elements_by_class_name("stream-item__date")
author = driver.find_elements_by_class_name("byline__author-name")
link = driver.find_elements_by_class_name("stream-item__title")
list_articles=[]
for t,d,da,a,l in zip(title,desc,date,author,link):
    list_articles.append({"title":t.text,"desc":d.text,"date":da.text,"author":"By "+a.text,"link":l.get_attribute("href")})
    
    
for a in list_articles:
    print(a)
    print("\n")
```
