```python
from selenium import webdriver
import pandas as pd
import bs4
```

```python
driver = webdriver.Chrome('C:/Users/My PC/Downloads/chromedriver')
driver.get("https://venturebeat.com/category/entrepreneur/")
```

```python
uid = driver.find_elements_by_tag_name("article")

for a in uid:
    print(a.text)
    href = a.find_element_by_tag_name("a")
    print(href.get_attribute("href"))
    print("-"*70)
```

```python
uid = driver.find_element_by_tag_name("article")
title = uid.find_elements_by_class_name("article-title")
for a in title:
    print(a.text)
    print("-"*70)
```
